```python
import fitz  # PyMuPDF
import numpy as np
import os
import pandas as pd
import re
import string
import nltk
from nltk.corpus import stopwords
from PyPDF2 import PdfReader
from sklearn.feature_extraction.text import CountVectorizer

agg_data = pd.read_csv("agg_data4.csv")
agg_data.head()
```

First we extract the text from each pdf and put create a new column in agg_data.

```python
def extract_text_from_pdf(pdf_path):
    try:
        pdf_document = fitz.open(pdf_path)
        text = ""
        for page_num in range(len(pdf_document)):
            page = pdf_document.load_page(page_num)
            text += page.get_text()
        return text
    except Exception as e:
        print(f"Error reading {pdf_path}: {e}")
        return ""

# Pdf directory
current_directory = os.getcwd()
folder_path = os.path.join(current_directory, 'pdf_reports')

agg_data['pdf_text_raw'] = agg_data['pdf'].apply(lambda pdf: extract_text_from_pdf(os.path.join(folder_path, pdf)))
```

We create another column, with the cleaned text.

```python
def preprocess_text(text):
    text = text.lower()  # Lowercase
    text = re.sub(r'\d+', '', text)  # Removing numbers
    text = text.replace('\n', '')  # Removing new line characters
    text = text.translate(str.maketrans('', '', string.punctuation))  # Removing punctuation
    return text

agg_data['pdf_text_cleaned'] = agg_data['pdf_text_raw'].apply(preprocess_text)
```

On the raw text, we extract the following features:
- word_count: total number of words in pdf
- character_count: total number of characters in pdf
- word_density: average length of words used
- sentence count
- average sentence length
- count of unique words
- count of used digits
- count of punctuation used
- number of pages in pdf

```python
def word_count(text):
    words = text.split()
    return len(words)

def character_count(text):
    return len(text)

def word_density(text):
    words = text.split()
    if len(words) == 0:
        return 0
    return sum(len(word) for word in words) / len(words)

def sentence_count(text):
    sentences = re.split(r'[.!?]+', text)
    return len([s for s in sentences if s.strip()])

def avg_sentence_length(text):
    sentences = re.split(r'[.!?]+', text)
    words = text.split()
    if len(sentences) == 0:
        return 0
    return len(words) / len(sentences)

def unique_word_count(text):
    words = text.split()
    unique_words = set(words)
    return len(unique_words)

def digit_count(text):
    return sum(c.isdigit() for c in text)

def punctuation_count(text):
    return sum(1 for char in text if char in string.punctuation)

agg_data['word_count'] = agg_data['pdf_text_raw'].apply(word_count)
agg_data['character_count'] = agg_data['pdf_text_raw'].apply(character_count)
agg_data['word_density'] = agg_data['pdf_text_raw'].apply(word_density)
agg_data['sentence_count'] = agg_data['pdf_text_raw'].apply(sentence_count)
agg_data['avg_sentence_length'] = agg_data['pdf_text_raw'].apply(avg_sentence_length)
agg_data['unique_word_count'] = agg_data['pdf_text_raw'].apply(unique_word_count)
agg_data['digit_count'] = agg_data['pdf_text_raw'].apply(digit_count)
agg_data['punctuation_count'] = agg_data['pdf_text_raw'].apply(punctuation_count)
```

Before moving on with feature engineering, we separate the complete dataset in training, testing and validation. We use train and test for feature engineering and the validation set for final model evaluation. Since we have a limited amount of observations, we use only 10% for validation purposes:

```python
validation_data = agg_data.sample(frac=0.1, random_state = 123)
train_test_data = agg_data.drop(validation_data.index)
train_test_data.shape, validation_data.shape
```

We create several functions for text analysis:

- For different ratings, we inspect which words that could imply an assessment, are most often
- We decide against a tf-idf matrix because semantic cannot be checked

```python
# Helper Function for removing german stopwords
def stopwords(text, remove_stopwords=True):    
    if remove_stopwords == True:
        stopwords_german = set(stopwords.words('german'))
        words = text.split()
        filtered_words = [word for word in words if word.lower() not in stopwords_german]

        text = ' '.join(filtered_words)
    else:
        pass

    return text

# Function for reading project pdfs with a certain score
def read_files(agg_data, score, remove_stopwords=True):
    df = agg_data[agg_data["overall_rating"].isin(score)]
    pdfs = list(df["pdf_text_cleaned"])
    return pdfs

def most_used_words(arrays, k):
    # Finds the words that are most frequent in texts
    
    # 1. Creates a document term matrix by stacking all arrays and calcaulate mean values for every column
    stacked_arrays = np.vstack(arrays)
    mean_values = np.mean(stacked_arrays, axis=0)
    
    # 2. Sorting indices by mean values in descending order
    sorted_idx = np.argsort(-mean_values)
    highest_idx = sorted_idx[:k]
    return highest_idx, mean_values
```

We try to identify token-based features by looking for similarities in word use for different rating groups. Single words, bigrams and trigrams are looked at.

```python
# First we use this function to exract all texts for a specific rating
result = read_files(train_test_data, [1,2], remove_stopwords=False)

# Bag of Words
vectorizer = CountVectorizer(ngram_range=(3, 3))
X = vectorizer.fit_transform(result)

# List of arrays for similar_features function
arrays = [row for row in X.toarray()]

# Extracting the words that are used most often
idx_mean, most = most_used_words(arrays, 100)

feature_names = vectorizer.get_feature_names_out()

print("Most Frequently Used Words:")
print(feature_names[idx_mean])
print(" ")
print("Frequency of Most Used Words:")
print(most[idx_mean])
```
