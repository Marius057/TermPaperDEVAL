In this file, we:

- Use our project_data.csv to download every pdf

**Result: pdf_reports directory**

```python
import pandas as pd

data = pd.read_csv("project_data.csv")
```

Download every pdf in the dataframe and place them in a local directory:

```python
import os
import pandas as pd
import requests

current_directory = os.getcwd()
save_directory = os.path.join(current_directory, 'pdf_reports')

if not os.path.exists(save_directory):
    os.makedirs(save_directory)

# Download PDF function
def download_pdf(url, save_directory):
    try:
        response = requests.get(url)
        response.raise_for_status()
        filename = os.path.join(save_directory, os.path.basename(url))
        with open(filename, 'wb') as file:
            file.write(response.content)
        print(f"Downloaded: {filename}")
    except requests.exceptions.RequestException as e:
        print(f"Failed to download {url}: {e}")

# Download every PDF in data
for url in data['url']:
    download_pdf(url, save_directory)
```
